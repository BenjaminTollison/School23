```python
def sonicMach(LocalMach:float,gamma: float):
    M = LocalMach
    mach_number_squared = ((gamma+1)*M**2)/(2+(gamma-1)*M**2)
    return float(np.sqrt(mach_number_squared))
test_cases = np.linspace(.001,3,20)
for i in test_cases:
    print("M_local = {} and M* = {}".format(round(i,5) , round(sonicMach(i,1.4),5)))
```

```python
densityRatio = lambda M , gamma: (1+((gamma-1)/2)*M**2)**(-1/(gamma-1))
M_values = np.linspace(0,1,100)
plt.plot(M_values,densityRatio(M_values,1.4),label='rho/rho_0')
plt.xlabel("M")
plt.ylabel('rho/rho_0')
plt.hlines(0.95,0,1,linestyles='dashed')
plt.vlines(0.32,0,1,linestyles='dashed')
plt.show()
```
