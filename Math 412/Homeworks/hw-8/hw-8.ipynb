```python
### Benjamin Tollison ###
import sympy as sp
from IPython.display import display, Math, Latex
from sympy.solvers.solveset import linsolve
from sympy.solvers.pde import pdsolve
from sympy import  lambdify, Matrix,sin,cos, sinh, cosh, exp, pi, symbols, integrate, Function, Eq
import numpy as np
import matplotlib.pyplot as plt
def displayEquations(LHS,RHS):
    left = sp.latex(LHS)
    right = sp.latex(RHS)
    display(Math(left + '=' + right))
    np.set_printoptions(suppress=True)
```

```python
first_plot = plt.figure().add_subplot(projection='3d')
t = np.linspace(0,1,100)
x = np.linspace(0,np.pi,100)
u_xt = 3*np.exp(-9*t)*np.sin(3*x)
first_plot.plot(x,t,u_xt,label='u(x,t)')
first_plot.set_title('u(x,t)')
first_plot.set_xlim(0, np.pi)
first_plot.set_ylim(0, 1)
first_plot.set_xlabel('x')
first_plot.set_ylabel('t')
first_plot.set_zlabel('u(x,t)')
plt.show()
```
