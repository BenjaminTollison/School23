## *raging hell of VARIABLES* ⚡



```python
M = 0.52 # Mach Number
Uinf = 503.36 # ft/s
rho = 0.000738
qinf = 0.5*rho*Uinf**2
print('qinf =', qinf)

g = 32.2 # ft/s^2
interferencefactorF = 1.0

Ain = 15 # sq ft
W = 22000 #lb
vol = 1000 # ft^3
Xcg = 27 # ft
Zcg = 0 # ft
m = W/32.1745 # kg

# Tail efficiency factors
nH = 0.5
nV = 1

qH = nH * qinf
deda = 0.75
dodb = 0.12

# Lift curve slopes for 3D
ClaW = 5.6
ClaH = 6
ClaV = 6

# Surface Areas
SW = 400 # ft^2
SH = 100 # ft^2
SV = 70 # ft^2

# Aspect ratios
ARW = 3.6
ARH = 3.4
ARV = 1.5

# Spans
bW = np.sqrt(ARW * SW) # ft
bH = np.sqrt(ARH * SH) # ft
bV = np.sqrt(ARV * SV) # ft
print(f'\nSpans: \nWing = {bW}, H-tail = {bH}, V-tail = {bV}')

# Root chords
crW = 17 # ft
crH = 10 # ft
crV = 10 # ft

# Taper ratios
ratW = (2*bW - (crW*ARW))/(crW*ARW)
ratH = (2*bH - (crH*ARH))/(crH*ARH)
ratV = (2*bV - (crV*ARV))/(crV*ARV)
print(f'\nTaper ratios: \nWing = {ratW}, H-tail = {ratH}, V-tail = {ratV}')

# c bars
cwbar = (2*(1 + ratW + ratW**2)/(3*(1 + ratW))) * crW
chbar = (2*(1 + ratH + ratH**2)/(3*(1 + ratH))) * crH
cvbar = (2*(1 + ratV + ratV**2)/(3*(1 + ratV))) * crV
print(f'\ncbars: \nWing = {cwbar}, H-tail = {chbar}, V-tail = {cvbar}')

# Quard chord sweep
sweep_quartW = np.radians(45)
sweep_quartH = np.radians(45)
sweep_quartV = np.radians(45)

# Leading edge sweep
def sweep_edge(AR, ratio, sweep):
  pos = (1 - ratio)/(1 + ratio)
  return np.arctan((AR * np.tan(sweep) + pos)/AR)

sweep_LEW = sweep_edge(ARW, ratW, sweep_quartW)
sweep_LEH = sweep_edge(ARH, ratH, sweep_quartH)
sweep_LEV = sweep_edge(ARV, ratV, sweep_quartV)
print(f'\nLeading Edge Sweep Angles: \nWing = {np.degrees(sweep_LEW)}, H-tail = {np.degrees(sweep_LEH)}, V-tail = {np.degrees(sweep_LEV)}')


# X LE MAC
def x_LE(ratio, sweep_LE, b):
  pos = (1 + 2*ratio)/(3 * (1 + ratio))
  return (b/2)*pos*np.tan(sweep_LE)

x_LEW = x_LE(ratW, sweep_LEW, bW)
x_LEH = x_LE(ratH, sweep_LEH, bH)
x_LEV = x_LE(ratV, sweep_LEV, bV)
print('\nX LE MAC:', x_LEW, x_LEH, x_LEV)

# X apex
XapexW = 16 # ft
XapexH = 36 # ft
XapexV = 35 # ft

XACW = XapexW + x_LEW + 0.32*cwbar
print('\nXACW =', XACW)
XACWbar = XACW/cwbar
print('XACWbar =', XACWbar)

XACWF = XapexW + x_LEW + 0.27*cwbar
print('\nXACWF =', XACWF)
XACWFbar = XACWF/cwbar
print('XACWFbar =', XACWFbar)

XACH = XapexH + x_LEH + 0.32*chbar
print('\nXACH=', XACH)
XACHbar = XACH/cwbar
print('XACHbar =', XACHbar)

XACV = XapexV + x_LEV + 0.32*cvbar
print('\nXACV=', XACV)
XACVbar = XACV/cvbar
print('XACVbar =', XACVbar)

# Half Sweep Angle
def half_sweep(sweep, ratio, AR):
  pos = (1 - ratio)/(1 + ratio)
  num = AR*np.tan(sweep)
  return np.arctan((num - 4*(0.5 - 0.25)*pos)/AR)

sweep_halfW = half_sweep(sweep_quartW, ratW, ARW)
sweep_halfH = half_sweep(sweep_quartH, ratH, ARH)
sweep_halfV = half_sweep(sweep_quartV, ratV, ARV)
print(f'\nHalf Sweep Angles: \nWing = {np.degrees(sweep_halfW)}, H-tail = {np.degrees(sweep_halfH)}, V-tail = {np.degrees(sweep_halfV)}')

delXACHbar = (XACH - Xcg)/cwbar
print('\ndelXACHbar =', delXACHbar)
delXACWFbar = (Xcg - XACWF)/cwbar
print('delXACWFbar =', delXACWFbar)

Ixx = 7210 # slug-ft^2
Iyy = 31000 # slug-ft^2
Izz = 37000 # slug-ft^2
Ixz = 1000 # slug-ft^2
```

## *make airplane fly ✈*

```python
cla = (2*np.pi)/(np.sqrt(1 - M**2))
print('cla =', cla)

beta = np.sqrt(1 - M**2)
print('beta =', beta)

k = cla/(2*np.pi)
print('k =', k)

kW = ClaW/(2*np.pi)
kH = ClaH/(2*np.pi)
kV = ClaV/(2*np.pi)
print('k:', kW, kH, kV)

def CLa_func(AR, k, sweep):
  num = 2*np.pi* AR
  mult1 = (AR**2 * beta**2) / k**2
  mult2 = (1 + (((np.tan(sweep))**2) / beta**2))
  denom = 2 + np.sqrt((mult1 * mult2) + 4)
  return num/denom

CLaW = CLa_func(ARW, kW, sweep_halfW) # 3D
CLaH = CLa_func(ARH, kH, sweep_halfH)
#CLaV = CLa(ARH, kV, sweep_halfV)
print('\nCLa \'s:', CLaW, CLaH)


def CLa_overall():
  return CLaW + CLaH*nH*(SH/SW)*(1 - deda)

#CLa = CLa_func(ARW, k, sweep_halfW)
#print('CLa =', CLa)

CLa2 = CLa_overall() # real CL_alpha
print('CLa2 =', CLa2)


# def CLaWF():
#  return CLa - CLaH * nH * (SH/SW) * (1 - deda)

CLaWF = CLaW
print('CLaWF =', CLaWF)

VH = (SH/SW) * delXACHbar
def CMa():
  return CLaWF*delXACWFbar - CLaH*nH*VH*(1-deda)

CMa = CMa()
print('CMa =', CMa)

SM = -CMa/CLa2
print('\nSM =', SM)

def CMiH():
  return -CLaH*nH*VH

CMiH = CMiH()
print('CMiH =', CMiH)

def XNPbar():
  return (CLaWF*XACWFbar + CLaH*nH*(SH/SW)*XACHbar*(1 - deda))/CLa2

XNPbar = XNPbar()
print('XNPbar =', XNPbar)

CLiH = CLaH * nH * (SH/SW)
print('CLiH =', CLiH)
```

## *Trim Calculator* ⛺

```python
from scipy.optimize import fsolve
from sympy import (symbols, sin, cos, solve, pi)
import math
a_0 = symbols('a_0')
T_0 = symbols('T_0')
i_h = symbols('i_h')


def equations(variables):
    a_0, i_h, T_0 = variables

    e = deda * a_0
    a_H = a_0 + i_h - e
    L_wf = CLaW * a_0 * qinf * SW #CLaWF is the same as CLaW
    L_h = CLaH * a_H * qH * SH
    C_D0 = 0.0145 + 0.1*((CLaW*a_0)**2)
    D_0 = qinf * SW * C_D0
    F_N0 = 2 * qinf * Ain * (np.cos(a_0)**2) * np.sin(a_0)

    # Trim equations
    #X = T_0 - W*np.sin(a_0) + L_wf * np.sin(a_0) - D_0 *np.cos(a_0) + L_h*np.sin(a_0 - e)
    X = T_0 * np.cos(a_0) - F_N0 * np.sin(a_0)- D_0 - L_h * np.sin(e)

    #Z = W*np.cos(a_0) - F_N0 - L_wf * np.cos(a_0) - D_0 * np.sin(a_0) - L_h* np.cos(a_0 - e)
    Z = W - T_0 * np.sin(a_0) - F_N0 * np.cos(a_0) - L_wf - L_h* np.cos(e)

    #M = F_N0 * (X_cg - 0) - L_wf * (X_ac_wf - X_cg)*np.cos(a_0) - D_0 * (X_ac_w - X_cg) * np.sin(a_0) - L_h * (X_ac_h - X_cg) * np.cos(a_0 - e)
    M = F_N0 * (Xcg - 0) - L_wf * (XACWF - Xcg)*np.cos(a_0) - D_0 * (XACWF - Xcg) * np.sin(a_0) - L_h * (XACH - Xcg) * np.cos(a_0 - e)

    return [X, Z, M]

# Initial guess for the variables
initial_guess = [np.radians(10), np.radians(-5), 2000]  # Provide your initial guess here
radtodeg = 180/np.pi
# Solve the equations
solution = fsolve(equations, initial_guess)

print('Alpha trim is ', solution[0]* radtodeg)
print('\nIncidence Angle is ', (solution[1]) * radtodeg)
print('\nTrim Thrust is ', solution[2])

alpha = solution[0]
i_h = solution[1]
thrust = solution[2]
e = deda * alpha
```

## *Every thing dependent on trim* ⛳

```python
def CL0():
  return CLaWF*alpha + nH*(SH/SW)*CLaH*(alpha)*np.cos(e) # alpha + deda + i_h

CL0 = CL0()
print('CL0 =', CL0)

def CD():
  return 0.0145 + 0.1*((CLaW*alpha)**2)
CD0 = CD()
print('CD0 =', CD0)

CDa = 0.2*CLaW**2 *alpha
print('CDa =', CDa)

FN = 2 * qinf * Ain * np.cos(alpha) ** 2 * np.sin(alpha)
print('\nFN =', FN)

prop_CP_M0=(FN*(Xcg-0))/(qinf*cwbar*SW)
print('Propulsive force =', prop_CP_M0)

CM0 = -prop_CP_M0
print('CM0 =', CM0)
```

## *Longitudinal Stability Derivatives* 🧮

```python
def CoeffDragDerivative(c): # also works for CLM
  num = M * np.cos(sweep_quartW)**2
  denom = 1 - (M**2 * np.cos(sweep_quartW)**2)
  CM = (num*c)/denom
  return CM

CDM = CoeffDragDerivative(CD0)
print('CDM =', CDM)

CLM = CoeffDragDerivative(CL0)
print('CLM =', CLM)

def CoeffMomentq(mult2): # also works for alpha dot
  mult1 = (XACH - Xcg)
  mult3 = np.cos(alpha)**2
  return -nH * (SH/SW) * CLaH * mult1 * mult2 * mult3 * (1/(Uinf*cwbar))

CMq = CoeffMomentq((XACH - Xcg))
print('CMq =', CMq)
CMadot = CoeffMomentq(deda*(XACH - XACW))
print('CMadot =', CMadot)

# CDU for XU
D_0 = qinf * SW * CD0
CDU = (2*D_0*Uinf)/(rho*SW*Uinf**(3/2))
print('\nCDU =', CDU)
# CLU for ZU
CLU = (2*W*Uinf)/(rho*SW*Uinf**(3/2))
print('CLU =', CLU)

def XZ(v, c):
    num = (v + c) *qinf* SW
    return num/m

def Mom(v):
  return (v*qinf*SW*cwbar)/Iyy

XU = -XZ(CD0, CD0)/Uinf
print('\nXU =', XU)
ZU = -XZ(CL0, CL0)/Uinf
print('ZU =', ZU)
MU = Mom(-(0 + (2/Uinf)*CM0))
print('MU =', MU)

Xa = XZ(-CDa, CL0)
print('\nXa =', Xa)
Za = -XZ(CLa2, CD0)
print('Za =', Za)
Ma = Mom(CMa)
print('Ma =', Ma)

Mq = Mom(CMq)
print('Mq =', Mq)
Madot = Mom(CMadot)
print('Madot =', Madot)
```

## *fyuuuuugoid* ⏰#

```python
A = np.array([[XU, Xa, 0, -32.17],
              [ZU/Uinf, Za/Uinf, 1, 0],
              [MU + (Madot*ZU)/Uinf, Ma + (Madot*Za)/Uinf, Mq + Madot, 0],
              [0 , 0, 1, 0]])

s1, s2 = eig(A)
for x in range(len(s1)):
  print('E-value:', s1[x])

def flight (r):
  real = r.real
  print('Real =', real)
  imag = r.imag
  print('Imaginary =', imag)
  w_n = np.sqrt(real**2 + imag**2)
  print('Natural frequncy =', w_n)
  damp = -real/w_n
  print('Damping ratio =', damp)
  w = w_n * np.sqrt(1 - damp**2)
  print('Damped Frequency =', w)
  t_p = (2*np.pi)/w
  print('Time of period =', t_p)
  T_C = 1/abs(real)
  print('Time constant =', T_C)
  t_half_dub = np.log(2)/np.abs(real)
  if t_half_dub < 0:
    print('Time to half =', t_half_dub)
  else:
    print('Time to double =', t_half_dub)
  cycle_half = t_half_dub/t_p
  print('Cycles to half =', cycle_half)
  return w

print('\nShort period -- For the eigenvalue:', s1[0])
sp_omega = flight(s1[0])
print('\nPhugoid -- For the eigenvalue:', s1[2])
flight(s1[2])

n_div_a = -Za/g
print('\nn/a =', n_div_a)
CAP = sp_omega**2/n_div_a
print('CAP =', CAP)
```
