```python
# This program is to see how many transactions it takes for the US goverment to take all of your money #
# sales tax set to %6.25 which is texas
# avg wage salary is set to 49,949
# Seeing how fast a dollar gets 'repo'
import numpy as np
import matplotlib.pyplot as plt
    # Texas tax rate
def transaction_counter(intial_dollar_amount):
    transaction_counter = 0
    while intial_dollar_amount >= .009:
        sales_tax = intial_dollar_amount * .0625
        after_wage_tax = intial_dollar_amount * .7912
        intial_dollar_amount = after_wage_tax - sales_tax
        transaction_counter += 1
    return transaction_counter
list_of_dollars = np.linspace(1,1000000,1000)
number_of_transactions = []
for i in list_of_dollars:
    plt.scatter((i),(transaction_counter(i)))
    number_of_transactions.append(transaction_counter(i))
plt.xlabel('Initial Dollar Amount \n *log10 scale*')
plt.xscale('log')
plt.ylabel('Number of Transactions')
plt.grid()
plt.title('US Government to repo all Texas money')
plt.show()
print("$1 dissappears in {} transactions".format(np.min(number_of_transactions)))
print("${} dissappears in {} transactions".format(list_of_dollars[-1], np.max(number_of_transactions)))
```
